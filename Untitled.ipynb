```python
import pandas as pd

df = pd.read_csv(
    "s3://demandcast-sajan-eu-west-2/exports/foods_1_020_csv/20250819_230044_00023_nk4cd_f090c06e-cb22-4e8b-b0c3-a8c288e3194f.gz",
    compression="gzip",
    header=None  
)

df.columns = ["date", "sales", "lag_7", "roll_7", "sell_price"]

print(df.head())
```

```python
import numpy as np

df = df.replace("\\N", np.nan)

# Convert to proper dtypes
df["sales"] = df["sales"].astype(int)
df["lag_7"] = pd.to_numeric(df["lag_7"], errors="coerce")
df["roll_7"] = df["roll_7"].astype(float)
df["sell_price"] = df["sell_price"].astype(float)

# Convert date column to datetime
df["date"] = pd.to_datetime(df["date"])

print(df.dtypes)
print(df.head(10))
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# Last 28 days = test set
horizon = 28
train_df = df.iloc[:-horizon]
test_df  = df.iloc[-horizon:]

features = ["lag_7", "roll_7", "sell_price"]
target = "sales"

X_train = train_df[features]
y_train = train_df[target]

X_test = test_df[features]
y_test = test_df[target]

train_set = lgb.Dataset(X_train, label=y_train)
params = {
    "objective": "regression",
    "metric": "rmse",
    "verbosity": -1,
    "boosting_type": "gbdt",
    "learning_rate": 0.1,
    "num_leaves": 31,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "seed": 42
}

model = lgb.train(
    params,
    train_set,
    num_boost_round=100
)

y_pred = model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"RMSE over last {horizon} days: {rmse:.4f}")

results = test_df[["date", "sales"]].copy()
results["predicted_sales"] = y_pred

print(results.head(10))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(results["date"], results["sales"], label="Actual")
plt.plot(results["date"], results["predicted_sales"], label="Predicted")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

horizon = 28
train_df = df.iloc[:-horizon].copy()
test_df  = df.iloc[-horizon:].copy()

# Choose features — start with the same as LightGBM, then try adding calendar features
base_features = ["lag_7", "roll_7", "sell_price"]
extra_features = ["day_of_week", "is_weekend"]  # comment out if you want exact parity
features = base_features + extra_features

# Drop rows with any NaNs in training features/target
train_df = train_df.dropna(subset=features + ["sales"])
test_df  = test_df.dropna(subset=features + ["sales"])  # usually already okay by the end

X_train, y_train = train_df[features], train_df["sales"]
X_test,  y_test  = test_df[features],  test_df["sales"]
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
rmse = mean_squared_error(y_test, y_pred_lr, squared=False)
mae  = mean_absolute_error(y_test, y_pred_lr)
mape = np.mean(np.abs((y_test - y_pred_lr) / (y_test + 1e-5))) * 100  # avoid /0

print(f"Linear Regression — RMSE: {rmse:.4f}  MAE: {mae:.4f}  MAPE: {mape:.2f}%")
```

```python
import numpy as np
import pandas as pd

# ----------------------------
# 1. Enrich Features
# ----------------------------

# Extra lags
df["lag_14"] = df["sales"].shift(14)
df["lag_28"] = df["sales"].shift(28)

# Extra rolling windows
df["roll_28"] = df["sales"].rolling(28).mean()
df["roll_56"] = df["sales"].rolling(56).mean()

# Calendar features
df["day_of_week"] = df["date"].dt.dayofweek   # 0=Mon, 6=Sun
df["month"] = df["date"].dt.month
df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)

# Price dynamics
df["price_diff_7"] = df["sell_price"] - df["sell_price"].shift(7)

# Drop rows with NaNs (first 56 days will be NaN due to rolling)
df = df.dropna().reset_index(drop=True)

# ----------------------------
# 2. Train / Test Split
# ----------------------------
horizon = 28
train_df = df.iloc[:-horizon]
test_df  = df.iloc[-horizon:]

features = [
    "lag_7", "lag_14", "lag_28",
    "roll_7", "roll_28", "roll_56",
    "sell_price", "price_diff_7",
    "day_of_week", "month", "is_weekend"
]
target = "sales"

X_train, y_train = train_df[features], train_df[target]
X_test, y_test = test_df[features], test_df[target]

# ----------------------------
# 3. Train LightGBM Model
# ----------------------------
import lightgbm as lgb
from sklearn.metrics import mean_squared_error

train_set = lgb.Dataset(X_train, label=y_train)
params = {
    "objective": "regression",
    "metric": "rmse",
    "verbosity": -1,
    "boosting_type": "gbdt",
    "learning_rate": 0.05,
    "num_leaves": 64,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "seed": 42
}

model = lgb.train(params, train_set, num_boost_round=200)

# ----------------------------
# 4. Predict & Evaluate
# ----------------------------
y_pred = model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"RMSE with richer features over last {horizon} days: {rmse:.4f}")

# Compare predictions
results = test_df[["date", "sales"]].copy()
results["predicted_sales"] = y_pred

print(results.head(10))

# ----------------------------
# 5. Feature Importance
# ----------------------------
lgb.plot_importance(model, max_num_features=15)
```
