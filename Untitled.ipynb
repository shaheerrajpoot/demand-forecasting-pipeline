```python
# === Feature engineering (no price columns) ===
import pandas as pd
import numpy as np

def add_time_features(df):
    df["dow"] = df["date"].dt.weekday
    df["dom"] = df["date"].dt.day
    df["week"] = df["date"].dt.isocalendar().week.astype(int)
    df["quarter"] = df["date"].dt.quarter
    df["is_month_start"] = df["date"].dt.is_month_start.astype(int)
    df["is_month_end"] = df["date"].dt.is_month_end.astype(int)
    return df

def add_sales_lags_rolls(df):
    df = df.sort_values(["id","date"])
    for L in (7, 14, 28):
        df[f"lag_{L}"] = df.groupby("id")["sales"].shift(L).astype(float)
    # rolling on shifted sales to avoid leakage
    for W in (7, 28, 56):
        df[f"roll_{W}"] = df.groupby("id")["sales"].transform(lambda s: s.shift(1).rolling(W).mean())
    return df

def add_snap_flags(df, state_col="state_id"):
    # If calendar SNAP columns exist, derive a single 'snap' flag; otherwise default to 0
    cols = df.columns
    if {"snap_CA","snap_TX","snap_WI"}.issubset(cols):
        df["snap"] = 0
        df.loc[df[state_col]=="CA","snap"] = df.loc[df[state_col]=="CA","snap_CA"]
        df.loc[df[state_col]=="TX","snap"] = df.loc[df[state_col]=="TX","snap_TX"]
        df.loc[df[state_col]=="WI","snap"] = df.loc[df[state_col]=="WI","snap_WI"]
    elif "snap" in cols:
        pass
    else:
        df["snap"] = 0
    return df

def build_feature_matrix(df):
    df = df.copy()

    # dtypes
    if not np.issubdtype(df["date"].dtype, np.datetime64):
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

    # compose
    df = add_time_features(df)
    df = add_sales_lags_rolls(df)
    df = add_snap_flags(df)

    # simple calendar flags if available
    if "event_name_1" in df.columns:
        df["is_event"] = df["event_name_1"].notna().astype(int)
    else:
        df["is_event"] = 0
    df["is_weekend"] = (df["dow"] >= 5).astype(int)

    # final feature list (no price features)
    feature_cols = [
        "lag_7","lag_14","lag_28",
        "roll_7","roll_28","roll_56",
        "snap","is_event","is_weekend",
        "dow","dom","week","quarter","is_month_start","is_month_end"
    ]

    # burn-in removal
    df = df.dropna(subset=["lag_7","roll_7"]).reset_index(drop=True)
    return df, feature_cols
```

```python
import mlflow
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt

def train_with_mlflow(df, feature_cols, test_days=28, run_name="lgbm_run"):
    df = df.sort_values("date").dropna(subset=feature_cols + ["sales"])
    train = df.iloc[:-test_days]
    test  = df.iloc[-test_days:]
    Xtr, ytr = train[feature_cols], train["sales"]
    Xte, yte = test[feature_cols], test["sales"]

    params = dict(
        n_estimators=600, learning_rate=0.05, max_depth=-1,
        subsample=0.9, colsample_bytree=0.9, random_state=42
    )

    with mlflow.start_run(run_name=run_name):
        mlflow.log_params(params)
        mlflow.log_param("features", ",".join(feature_cols))
        mlflow.log_param("horizon_days", test_days)
        mlflow.set_tag("series_id", str(df["id"].iloc[0]))

        model = LGBMRegressor(**params)
        model.fit(Xtr, ytr)
        preds = model.predict(Xte)
        rmse = sqrt(mean_squared_error(yte, preds))
        mlflow.log_metric("rmse", rmse)

        # save predictions as artifact
        out = test[["date","id","sales"]].copy()
        out["pred"] = preds
        out_path = "artifacts/test_preds.csv"
        out.to_csv(out_path, index=False)
        mlflow.log_artifact(out_path)

    return rmse

rmse = train_with_mlflow(g_feat, feat_cols, test_days=28, run_name=f"LGBM_{SERIES}_h28")
print("Series:", SERIES)
print("RMSE:", rmse)
```

```python
from pathlib import Path
import pandas as pd

# Make sure artifacts dir exists (for summary CSV)
Path("artifacts").mkdir(exist_ok=True)

def train_many_lags_only(df_all, ids=None, test_days=28, max_ids=20):
    # pick a manageable set of series
    if ids is None:
        ids = df_all["id"].drop_duplicates().head(max_ids).tolist()

    results = []
    lag_feats = ["lag_7","lag_14","lag_28"]

    for sid in ids:
        try:
            g = df_all[df_all["id"] == sid].copy().sort_values("date")
            g_feat, _ = build_feature_matrix(g)
            # ensure we only use lags
            feats = [c for c in lag_feats if c in g_feat.columns]
            if len(feats) < 2:  # need at least some lags
                results.append({"id": sid, "rmse": None, "note": "insufficient lag features"})
                continue

            rmse = train_with_mlflow(
                g_feat, feats, test_days=test_days, run_name=f"LGBM_{sid}_lags_only"
            )
            results.append({"id": sid, "rmse": rmse})
            print(f"Done: {sid}  RMSE={rmse:.4f}")
        except Exception as e:
            results.append({"id": sid, "rmse": None, "error": str(e)})
            print(f"Error: {sid} -> {e}")

    df_res = pd.DataFrame(results).sort_values(
        by=["rmse"], ascending=[True], na_position="last"
    )
    out_path = "artifacts/multi_series_rmse_lags_only.csv"
    df_res.to_csv(out_path, index=False)
    print(f"\nSaved summary -> {out_path}")
    return df_res

# Kick off for first 20 series (adjust max_ids if you want more/less)
multi_results = train_many_lags_only(df_enriched, max_ids=20, test_days=28)
multi_results.head(10)
```

```python
def build_feature_matrix(df):
    df = df.copy()

    # dtypes
    if not np.issubdtype(df["date"].dtype, np.datetime64):
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

    # time/calendar features
    df = add_time_features(df)
    df = add_sales_lags_rolls(df)
    df = add_snap_flags(df)

    # event flag
    if "event_name_1" in df.columns:
        df["is_event"] = df["event_name_1"].notna().astype(int)
    else:
        df["is_event"] = 0
    df["is_weekend"] = (df["dow"] >= 5).astype(int)

    # --- NEW FEATURES ---
    # rolling std (volatility of sales)
    df["roll_std_28"] = (
        df.groupby("id")["sales"].transform(lambda s: s.shift(1).rolling(28).std())
    )

    # cumulative sales trend
    df["cum_sales"] = df.groupby("id")["sales"].cumsum()

    # final feature list
    feature_cols = [
        "lag_7","lag_14","lag_28",
        "roll_7","roll_28","roll_56","roll_std_28",
        "snap","is_event","is_weekend",
        "dow","dom","week","quarter","is_month_start","is_month_end",
        "cum_sales"
    ]

    df = df.dropna(subset=["lag_7","roll_7"]).reset_index(drop=True)
    return df, feature_cols
```

```python
# Extend build_feature_matrix with lag ratio & momentum features
def build_feature_matrix(df):
    df = df.copy()

    # dtypes
    if not np.issubdtype(df["date"].dtype, np.datetime64):
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

    # time/calendar features
    df = add_time_features(df)
    df = add_sales_lags_rolls(df)
    df = add_snap_flags(df)

    # event flag
    if "event_name_1" in df.columns:
        df["is_event"] = df["event_name_1"].notna().astype(int)
    else:
        df["is_event"] = 0
    df["is_weekend"] = (df["dow"] >= 5).astype(int)

    # volatility & trend
    df["roll_std_28"] = df.groupby("id")["sales"].transform(lambda s: s.shift(1).rolling(28).std())
    df["cum_sales"]   = df.groupby("id")["sales"].cumsum()

    # --- NEW: lag vs rolling level features (ratios / momentum / z-score) ---
    eps = 1e-6
    df["lag7_over_roll28"]  = df["lag_7"]  / (df["roll_28"] + eps)
    df["lag14_over_roll28"] = df["lag_14"] / (df["roll_28"] + eps)
    df["lag28_over_roll28"] = df["lag_28"] / (df["roll_28"] + eps)  # ~1 when stable

    df["mom_7_28"]  = df["lag_7"]  - df["roll_28"]
    df["mom_14_28"] = df["lag_14"] - df["roll_28"]
    df["mom_28_28"] = df["lag_28"] - df["roll_28"]

    # z-score of last week vs monthly mean (guards against division by zero)
    df["z_7_28"] = (df["lag_7"] - df["roll_28"]) / (df["roll_std_28"].fillna(0) + eps)

    # final feature list
    feature_cols = [
        # core lags & rolls
        "lag_7","lag_14","lag_28",
        "roll_7","roll_28","roll_56","roll_std_28",
        # ratios / momentum / z
        "lag7_over_roll28","lag14_over_roll28","lag28_over_roll28",
        "mom_7_28","mom_14_28","mom_28_28","z_7_28",
        # calendar
        "snap","is_event","is_weekend",
        "dow","dom","week","quarter","is_month_start","is_month_end",
        # trend
        "cum_sales",
    ]

    # burn-in removal
    df = df.dropna(subset=["lag_7","roll_7"]).reset_index(drop=True)
    return df, feature_cols

```
